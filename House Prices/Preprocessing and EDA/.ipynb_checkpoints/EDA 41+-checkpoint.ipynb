# Importing Required Packages

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats
import statsmodels.api as sm

import statsmodels.graphics.tsaplots as sgt

import warnings
warnings.filterwarnings("ignore")
```

# Importing Datasets

```python
backup_train = pd.read_csv("../Data/train.csv")
backup_test = pd.read_csv("../Data/test.csv")

train_df = backup_train.copy()
test_df = backup_test.copy()
```

```python
test_df.info()
```

```python
train_df.info()
```

# Helper Functions

```python
def scatter_plots(df, column):
    cat_column = list(df[column].unique())
    
    for cat in cat_column:
        plt.scatter(x = df[df[column] == cat].Id, 
                    y = df[df[column] == cat].SalePrice, 
                    label = cat)

    plt.xlabel(column)
    plt.ylabel("SalePrice")

    plt.legend(bbox_to_anchor = (1.25, 1))
    plt.show()
```

```python
dates = pd.to_datetime(train_df["MoSold"].astype("str") + "-" + train_df["YrSold"].astype("str"), 
                       format = "%m-%Y")
mo_yr_df = pd.DataFrame({"Dates": dates, "SalePrice": train_df["SalePrice"]})
mo_yr_df.sort_values(by = "Dates", inplace = True) 

mo_yr_df.groupby(by = "Dates")["SalePrice"].median().plot()

plt.show()
```

**Heatmap between days past 2006 and SalePrice**

```python
dates = pd.to_datetime(train_df["MoSold"].astype("str") + "-" + train_df["YrSold"].astype("str"), 
                       format = "%m-%Y")

days_past_2006 = dates - pd.to_datetime("2006-01-01", format = "%Y-%m-%d")

days_past_df = pd.DataFrame({"Days": days_past_2006.dt.days, "SalePrice": train_df["SalePrice"]})

# days_past_2006.dt.days
sns.heatmap(days_past_df.corr(), annot = True)
plt.show()
```

# Helper Functions

```python
def scatter_plots(df, column):
    cat_column = list(df[column].unique())
    
    for cat in cat_column:
        plt.scatter(x = df[df[column] == cat].Id, 
                    y = df[df[column] == cat].SalePrice, 
                    label = cat)

    plt.xlabel(column)
    plt.ylabel("SalePrice")

    plt.legend(bbox_to_anchor = (1.25, 1))
    plt.show()
```

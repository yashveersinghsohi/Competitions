# Importing Required Packages

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats
import statsmodels.api as sm

from sklearn.feature_selection import SelectKBest, f_regression, mutual_info_regression
from sklearn.linear_model import LinearRegression

import statsmodels.graphics.tsaplots as sgt

import warnings
warnings.filterwarnings("ignore")
```

# Reading Data

```python
backup_train = pd.read_csv("../Data/train.csv")
backup_test = pd.read_csv("../Data/test.csv")

train_df = backup_train.copy()
test_df = backup_test.copy()
```

```python
categorical = [
    "Neighborhood", "HouseStyle", "OverallCond", "Exterior1st", 
    "Exterior2nd", "MasVnrType", "BsmtExposure", "BsmtFinType1", 
    "HeatingQC", "Fireplaces", "GarageType"
    # ExterCond(Low)
    # BsmtCond(Low)
]

numeric = [
    "LotFrontage", "LotArea", "MasVnrArea", "BsmtFinSF1"
]

time = ["GarageYrBlt"]

target = ["SalePrice"]

features = categorical + numeric + time

train_med_df = train_df[features]
# train_high_df["YearDiff"] = train_df["YrSold"] - train_df["YearBuilt"]
train_med_df["SalePrice"] = train_df["SalePrice"]

print(train_med_df.shape)

train_med_df.head()
```

### LotFrontage

**Some Null Values**

```python
df = train_df[["LotFrontage", "SalePrice"]]
df["LotFrontage"].isnull().sum()
```

**Distribution of SalePrice when LotFrontage is Null**

```python
df[df["LotFrontage"].isnull()]["SalePrice"].describe()
```

**Scatter Plots with and without outliers**

```python
fig, ax = plt.subplots(1, 2, figsize = (12, 4), sharey = True)

sns.scatterplot(x = "LotFrontage", y = "SalePrice", data = df, ax = ax[0])
sns.scatterplot(x = "LotFrontage", y = "SalePrice", data = df[df["LotFrontage"]<300], ax = ax[1])

ax[0].set_title("With Outliers", size = 24)
ax[1].set_title("Without Outliers", size = 24)

plt.show()
```

**Using Linear Regression to fill null values**

```python
# Predicting Null Values
X, y = df[df["LotFrontage"]<300]["SalePrice"].values, df[df["LotFrontage"]<300]["LotFrontage"].values

lr = LinearRegression().fit(X.reshape((-1, 1)), y)
print(f"Coefficient: {lr.coef_}\t Intercept: {lr.intercept_}")

null_vals = lr.predict(df[df["LotFrontage"].isnull()]["SalePrice"].values.reshape(-1, 1))

# Storing Null Values
df_null = df[df["LotFrontage"].isnull()]
df_null["Null_vals"] = null_vals

# Plotting Null Values
fig, ax = plt.subplots()

sns.scatterplot(x = "LotFrontage", y = "SalePrice", data = df[df["LotFrontage"]<300], ax = ax)
sns.scatterplot(x = "Null_vals", y = "SalePrice", data = df_null, color = "red", ax = ax, label = "Null Vals")

ax.set_xlabel("LotFrontage")

plt.legend()
plt.show()
```

**Heatmaps with and without Nulls**

```python
df = train_med_df[train_med_df["LotFrontage"]<300][["LotFrontage", "SalePrice"]]

df_null = train_med_df[["LotFrontage", "SalePrice"]]
null_idx = train_med_df[train_med_df["LotFrontage"].isnull()].index
df_null.at[null_idx, "LotFrontage"] = lr.predict(train_med_df.loc[null_idx, "SalePrice"].values.reshape((-1, 1)))

fig, ax = plt.subplots(1, 2, figsize = (12, 4))

sns.heatmap(df.corr(), annot = True, ax = ax[0])
ax[0].set_title("Without Nulls")

sns.heatmap(df_null.corr(), annot = True, ax = ax[1])
ax[1].set_title("With Nulls Filled")

plt.show()
```

**Distribution and QQ plots for Log transformed data**

```python
df = train_med_df[train_med_df["LotFrontage"]<300][["LotFrontage", "SalePrice"]]

df_null = train_med_df[["LotFrontage", "SalePrice"]]
null_idx = train_med_df[train_med_df["LotFrontage"].isnull()].index
df_null.at[null_idx, "LotFrontage"] = lr.predict(train_med_df.loc[null_idx, "SalePrice"].values.reshape((-1, 1)))

fig, ax = plt.subplots(2, 2, figsize = (12, 8))

sns.distplot(np.log(df["LotFrontage"]), ax = ax[0, 0])
sm.qqplot(np.log(df["LotFrontage"]), line = "q", ax = ax[0, 1])

sns.distplot(np.log(df_null["LotFrontage"]), ax = ax[1, 0])
sm.qqplot(np.log(df_null["LotFrontage"]), line = "q", ax = ax[1, 1])

plt.show()
```

**Scatter plots and Heatmaps with log transformed data**

```python
df = train_med_df[train_med_df["LotFrontage"]<300][["LotFrontage", "SalePrice"]]
df["LotFrontage"] = np.log(df["LotFrontage"])
df["SalePrice"] = np.log(df["SalePrice"])

df_null = train_med_df[["LotFrontage", "SalePrice"]]
null_idx = train_med_df[train_med_df["LotFrontage"].isnull()].index
df_null.at[null_idx, "LotFrontage"] = lr.predict(train_med_df.loc[null_idx, "SalePrice"].values.reshape((-1, 1)))
df_null["LotFrontage"] = np.log(df_null["LotFrontage"])
df_null["SalePrice"] = np.log(df_null["SalePrice"])


fig, ax = plt.subplots(2, 2, figsize = (12, 8))

sns.scatterplot(x = "LotFrontage", y = "SalePrice", data = df, ax = ax[0, 0])
sns.scatterplot(x = "LotFrontage", y = "SalePrice", data = df_null, ax = ax[0, 1])

sns.heatmap(df.corr(), annot = True, ax = ax[1, 0])
sns.heatmap(df_null.corr(), annot = True, ax = ax[1, 1])

plt.show()
```

### Neighborhood

**Creating a mapping dictionary**

```python
df = train_med_df[["Neighborhood", "SalePrice"]]
df_neighbor = pd.DataFrame(df.groupby(by = "Neighborhood")["SalePrice"].median().sort_values()).reset_index()


map_dict = dict()
for idx, neighbor in zip(df_neighbor.index, df_neighbor["Neighborhood"]):
    map_dict[neighbor] = idx
```

**Heatmap with this mapping dictionary**

```python
df["Neighborhood"] = df["Neighborhood"].map(map_dict)

sns.heatmap(df.corr(), annot = True)
plt.show()
```

**Highly correlated with SalePrice. Use this mapping dictionary**
